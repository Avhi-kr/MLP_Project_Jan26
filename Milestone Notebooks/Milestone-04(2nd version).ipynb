```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

# Load dataset
train_df = pd.read_csv('/kaggle/input/competitions/comment-category-prediction-challenge/train.csv')

# Fill NaN in comment column with empty string
train_df['comment'] = train_df['comment'].fillna('')

# Separate features and target
X = train_df.drop(columns=['label'])
y = train_df['label']

# ============ Part A: Stratified Split ============
X_train_a, X_val_a, y_train_a, y_val_a = train_test_split(
    X, y, test_size=0.4, random_state=2306, stratify=y
)

# Sorted class-wise counts
train_counts_a = y_train_a.value_counts().sort_index().values
val_counts_a = y_val_a.value_counts().sort_index().values

print("Part A - Stratified:")
print("Training label counts:", train_counts_a)
print("Validation label counts:", val_counts_a)

# ============ Part B: Non-Stratified Split ============
X_train_b, X_val_b, y_train_b, y_val_b = train_test_split(
    X, y, test_size=0.4, random_state=2306, stratify=None
)

train_counts_b = y_train_b.value_counts().sort_index().values
val_counts_b = y_val_b.value_counts().sort_index().values

print("\nPart B - Non-Stratified:")
print("Training label counts:", train_counts_b)
print("Validation label counts:", val_counts_b)

# ============ Part C: Validation Distribution Difference ============
total_val_samples = len(y_val_a)  # same size for both splits

proportions_a = val_counts_a / total_val_samples
proportions_b = val_counts_b / total_val_samples

abs_diff = np.abs(proportions_a - proportions_b)
max_abs_diff = np.max(abs_diff)

print("\nPart C:")
print("Stratified validation proportions:", proportions_a)
print("Non-stratified validation proportions:", proportions_b)
print("Absolute differences:", abs_diff)
print("Maximum absolute difference:", max_abs_diff)
```

```python
import re
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack

# Use the stratified split from Q1 Part A
X_train = X_train_a.copy()
X_test = X_val_a.copy()
y_train = y_train_a.copy()
y_test = y_val_a.copy()

# Step 1: Drop created_date
X_train = X_train.drop(columns=['created_date'])
X_test = X_test.drop(columns=['created_date'])

# Step 2: Separate text column
text_x_train = X_train['comment']
text_x_test = X_test['comment']
X_train = X_train.drop(columns=['comment'])
X_test = X_test.drop(columns=['comment'])

# Step 3: Tabular Preprocessing
categorical_columns = ["race", "religion", "gender", "disability"]
numerical_columns = ["post_id", "emoticon_1", "emoticon_2", "emoticon_3",
                      "upvote", "downvote", "if_1", "if_2"]

categorical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=True))
])

numerical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

column_transformer = ColumnTransformer(
    transformers=[
        ('cat', categorical_pipeline, categorical_columns),
        ('num', numerical_pipeline, numerical_columns)
    ],
    remainder='passthrough'
)

# Fit on train, transform both
x_train_tabular = column_transformer.fit_transform(X_train)
x_test_tabular = column_transformer.transform(X_test)

# Step 4: Text Cleaning
def normalize_text(text):
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\s+', '  ', text).strip()
    return text

# Handle any remaining NaN in text columns
text_x_train = text_x_train.fillna('')
text_x_test = text_x_test.fillna('')

text_x_train_norm = text_x_train.apply(normalize_text)
text_x_test_norm = text_x_test.apply(normalize_text)

# Step 5: TF-IDF Vectorization
tfidf = TfidfVectorizer(stop_words='english', max_features=5000)
tf_idf_train = tfidf.fit_transform(text_x_train_norm)
tf_idf_test = tfidf.transform(text_x_test_norm)

# Step 6: Combine Features
X_train_final = hstack([x_train_tabular, tf_idf_train])
X_test_final = hstack([x_test_tabular, tf_idf_test])

# Answer: Sum of all values
sum_all = X_train_final.sum()
print(f"Sum of all values of X_train_final: {sum_all:.3f}")
```

```python
from sklearn.decomposition import TruncatedSVD

# Fit TruncatedSVD on X_train_final
svd = TruncatedSVD(n_components=300, random_state=2306)
X_train_reduced = svd.fit_transform(X_train_final)
X_test_reduced = svd.transform(X_test_final)

print("X_train_reduced shape:", X_train_reduced.shape)
print("X_test_reduced shape:", X_test_reduced.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV

# Model setup
rf = RandomForestClassifier(random_state=2306)

# Parameter grid
param_distributions = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 15]
}

# RandomizedSearchCV
randomized_search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_distributions,
    n_iter=5,
    cv=3,
    random_state=2306,
    n_jobs=-1
)

randomized_search.fit(X_train_reduced, y_train)

print("Best parameters:", randomized_search.best_params_)
print("Best n_estimators:", randomized_search.best_params_['n_estimators'])
```

```python
from sklearn.ensemble import AdaBoostClassifier

# Train AdaBoost
ada_model = AdaBoostClassifier(n_estimators=50, random_state=2306)
ada_model.fit(X_train_reduced, y_train)

# Extract estimator errors
estimator_errors = ada_model.estimator_errors_
print("Estimator errors:", estimator_errors)
print("Length:", len(estimator_errors))

# Compute variance
variance = np.var(estimator_errors)
print(f"Variance of estimator errors: {variance:.4f}")
```

```python
# Train RandomForest
rf_model = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=2306)
rf_model.fit(X_train_reduced, y_train)

# Extract feature importances
importances = rf_model.feature_importances_

# Find index of max importance (smallest index if tie)
max_importance_index = np.argmax(importances)
print(f"Index of feature with maximum importance: {max_importance_index}")
print(f"Maximum importance value: {importances[max_importance_index]}")
```

```python
# Network structure
N = X_train_reduced.shape[1]  # Number of input features (300)
hidden_1 = 128
hidden_2 = 64
hidden_3 = 32
output = 4

# Count weights (excluding biases)
weights = (N * hidden_1) + (hidden_1 * hidden_2) + (hidden_2 * hidden_3) + (hidden_3 * output)
print(f"N (input features): {N}")
print(f"Total weights (excluding biases): {weights}")
```

```python
from sklearn.neural_network import MLPClassifier

# Train MLP
mlp_model = MLPClassifier(
    hidden_layer_sizes=(128, 64, 32),
    activation='relu',
    solver='adam',
    max_iter=5,
    batch_size=32,
    random_state=2306
)

mlp_model.fit(X_train_reduced, y_train)

# Get final loss
final_loss = mlp_model.loss_
print(f"Final loss after 5 iterations: {round(final_loss, 4)}")
```

```python
from sklearn.metrics import f1_score

# Model A: alpha=0.0001 (default)
mlp_a = MLPClassifier(
    hidden_layer_sizes=(100,),
    max_iter=5,
    random_state=2306,
    alpha=0.0001
)
mlp_a.fit(X_train_reduced, y_train)
y_pred_a = mlp_a.predict(X_train_reduced)
f1_a = f1_score(y_train, y_pred_a, average='macro')

# Model B: alpha=1.0
mlp_b = MLPClassifier(
    hidden_layer_sizes=(100,),
    max_iter=5,
    random_state=2306,
    alpha=1.0
)
mlp_b.fit(X_train_reduced, y_train)
y_pred_b = mlp_b.predict(X_train_reduced)
f1_b = f1_score(y_train, y_pred_b, average='macro')

abs_diff = abs(f1_a - f1_b)
print(f"Macro F1 Score (Model A, alpha=0.0001): {f1_a}")
print(f"Macro F1 Score (Model B, alpha=1.0): {f1_b}")
print(f"Absolute difference: {round(abs_diff, 4)}")
```

```python
from sklearn.metrics import confusion_matrix

# Use the MLP model from Q7 (already trained)
y_val_pred = mlp_model.predict(X_test_reduced)

# Confusion matrix
cm = confusion_matrix(y_test, y_val_pred)
print("Confusion Matrix:")
print(cm)

# Sum of off-diagonal elements
total_samples = len(y_test)
diagonal_sum = np.trace(cm)
off_diagonal_sum = cm.sum() - diagonal_sum

result = off_diagonal_sum / total_samples
print(f"\nSum of off-diagonal elements: {off_diagonal_sum}")
print(f"Total validation samples: {total_samples}")
print(f"Off-diagonal sum / total samples: {round(result, 4)}")
```
